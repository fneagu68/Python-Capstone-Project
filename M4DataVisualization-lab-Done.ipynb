# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
df=pd.read_sql_query(QUERY,conn)

import matplotlib as mpl
import matplotlib.pyplot as plt

df["ConvertedComp"].plot(kind='hist')
plt.title("ConvertedComp")
plt.xlabel("ConvertedComp")
plt.ylabel("Frequency")
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
df=pd.read_sql_query(QUERY,conn)

df["Age"].plot(kind='box')
plt.title("Age")
plt.ylabel("Years")
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df=pd.read_sql_query(QUERY,conn)

df[["Age", "WorkWeekHrs"]].plot(kind="scatter", x="Age", y="WorkWeekHrs")
plt.title("Age-WorkWeekHrs scatter plot")
plt.ylabel("WorkWeekHrs")
plt.xlabel("Age")

plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df = df.dropna()
norm_CodeRevHrs = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df.plot.scatter(x = 'CodeRevHrs', y = 'WorkWeekHrs', s = norm_CodeRevHrs* 2500 + 10, figsize=(12,6) )
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT * FROM DatabaseDesireNextYear
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
df1=df["DatabaseDesireNextYear"].value_counts()
df2=df1.head()
df2
explode_list = [0.07, 0, 0, 0, 0] # ratio for each continent with which to offset each wedge.
df2.plot(kind='pie',
        figsize=(14, 9),
        autopct='%1.2f%%', 
        startangle=45,    
        shadow=True,       
        labels=None,         # turn off labels on pie chart
        pctdistance=1.14,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
        explode=explode_list
        )
plt.title('The top 5 databases that respondents wish to learn next year', y=1.08)
plt.axis('equal')
plt.legend(labels=df2.index, loc='upper right')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df_age = pd.read_sql_query(QUERY,conn) 


# group respondents by age and apply median() function 
df_age = df_age.groupby('Age', axis=0).median()

# step 2: plot data
df_age[30:35].plot(kind='bar', figsize=(10, 6), stacked=True)

plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Hours') # add y-label to the plot
plt.title('Median Hours by Age') # add title to the plot

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
Where Age between 45 and 60
"""

df=pd.read_sql_query(QUERY,conn)
df=df.groupby(by='Age')['ConvertedComp'].median()
df.plot(x="Age", kind='line', title='The median ConvertedComp for all ages from 45 to 60', figsize=(14, 8))
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""

df=pd.read_sql_query(QUERY,conn)
df1=df.groupby(by="MainBranch")["MainBranch"].count()
df1.plot(kind='barh', figsize=(14, 8),
        title='Horizontal bar chart using column MainBranch.')
plt.show()
```
